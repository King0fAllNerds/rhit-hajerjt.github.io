```python
with open('./pokemon.txt') as f:
    lines = f.readlines()
```

```python
# remove new line
trainer = lines[0][:-1]
for line in lines[1:]:
    poke_name = line.split(" ")[0]
    
    # checks for spaces in pokemon name
    name = ""
    for frag in line.split(" "):
        if frag == "(Lv.":
            poke_name = name
            break;
        else:
            name = name + " " +frag
    
    # checks for sinnohan form ending in name
    if poke_name[-3:] == "(S)":
        poke_name = poke_name[:-3]
    
    # determines level
    level = ""
    count = 0
    for frag in line.split(" "):
        if frag == "(Lv.":
            level = level + line.split(" ")[count+1][:-1]
            break
        count+=1
        
    item = ""
    recording_item = False
    for frag in line.split(" "):
        if recording_item:
            item = item + " " + frag
        if frag == "" or frag == "/":
            break;
        if frag == "@":
            recording_item = True
            
    ability = ""
    curr_move = ""
    moves = []
    recording_ability = False
    recording_moves = False
    for frag in line.split(" "):
        if frag == "/" and recording_ability:
            recording_moves= True
            recording_ability = False
            
        if recording_ability and frag != "":
            ability = ability + " " + frag
            
        if frag == "/" and not recording_moves:
            recording_ability = True
            
        if recording_moves and frag != "" and (frag[-1:] == ',' or frag == line.split(" ")[len(line.split(" "))-1]):
            curr_move = curr_move + " " + frag
            moves.append(curr_move[:-1].strip())
            curr_move = ""
            
        elif recording_moves and frag != "" and frag != "/":
            curr_move = curr_move + " " + frag
          
    print(poke_name.strip() + ': {')
    print("\""+trainer+"\""+": {")
    print("level: " + level+",")
    print("ability: " + "\""+ability.strip()+"\",")
    print("item: " +"\""+item.strip()+"\",")
    print("moves: [\""+moves[0]+"\", \""+moves[1]+"\", \""+moves[2]+"\", \""+moves[3]+"\"]")
    print("},")
    print("},")
```

```python
f.close()
```
